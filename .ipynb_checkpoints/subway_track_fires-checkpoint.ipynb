```python
%matplotlib inline 

import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
track_fires = pd.read_csv("track-fires.csv")
track_fires["Date"] = pd.to_datetime(track_fires["Date"])
track_fires["Year"] = track_fires["Date"].apply(lambda date: date.year)
track_fires["Comments"] = track_fires['Comments'].str.replace('\n', '')
track_fires["Comments"] = track_fires['Comments'].str.replace('\t', '')
track_fires["Comments"] = track_fires['Comments'].str.replace('\r', '')
```
